```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
data = pd.read_csv('Merged_Data.csv')
```

```python
data.info()
```

```python
data
```

```python
data = data[:149]
```

```python
fig, ax1 = plt.subplots(figsize=(18, 8))
ax2 = ax1.twinx()
ln1 = ax1.plot(data['Date'], data['Active_cases_new'], color="green")
ln2 = ax2.plot(data['Date'], data['O3'], color="red")
ax1.set_xlabel('Time (2020 from Jan to May)', fontsize='15')
ax1.set_ylabel('New active covid cases', color='g', fontsize='15')
ax2.set_ylabel('Ozone Concentration', color='r', fontsize='15')
ax1.xaxis.set_major_locator(mdates.MonthLocator())
ax1.legend(ln1+ln2, ['New active covid cases', 'Ozone Concentration'], fontsize='15')
plt.show()
```
